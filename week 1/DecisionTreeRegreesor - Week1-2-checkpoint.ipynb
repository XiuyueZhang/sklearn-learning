```python
from sklearn.datasets import load_boston
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
```

```python
# cross_val_score 交叉验证
boston = load_boston()
regressor = DecisionTreeRegressor(random_state=0)
cross_val_score(regressor, boston.data, boston.target, cv=10
               ,scoring='neg_mean_squared_error'   #分类树默认是score，回归树默认是R平方
               )
```

```python
# 一维回归的图像绘制
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(1)
x = np.sort(5* rng.rand(80,1),axis=0)  #80行，1列，生成80个0-1之间的随机数 -- 0-5之间  -- sort排序 生成横坐标x的数据
y = np.sin(x).ravel()  #y从二维降为一维
```

```python
plt.figure()
plt.scatter(x,y,s=20, edgecolor='black', c='darkorange',label='data')
```

```python
y[::5] += 3 * (0.5 - rng.rand(16))    #加噪声，模拟真实数据  y每间隔5个，取出数字+3倍（0-5-随机数）
```

```python
plt.figure()
plt.scatter(x,y,s=20, edgecolor='black', c='darkorange',label='data')
```

```python
regr_1 = DecisionTreeRegressor(max_depth=2)  
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(x, y)
regr_2.fit(x, y)
```

```python
x_test = np.arange(0.0,5.0,0.01)[:,np.newaxis] #np.newaxis增维成二维
y_1 = regr_1.predict(x_test)
y_2 = regr_2.predict(x_test)
```

```python
y_1
```

```python
plt.figure()
plt.scatter(x, y, edgecolor='black', c='darkorange',label='data')
plt.plot(x_test, y_1, color='cornflowerblue',label='max_depth=2',linewidth=2)
plt.plot(x_test, y_2, color='yellowgreen',label='max_depth=5',linewidth=2)
plt.xlabel('data')
plt.ylabel('target')
plt.title('Desicion Tree Regressor')
plt.legend()
plt.show()
```
