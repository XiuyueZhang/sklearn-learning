```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from time import time
import datetime
from sklearn.model_selection import train_test_split
```

```python
data = load_breast_cancer()
x = data.data
y = data.target
```

```python
x.shape
```

```python
np.unique(y) #去重 - 二分类
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
# 降维可视化
from sklearn.decomposition import PCA
X_dr = PCA(2).fit_transform(x)
```

```python
X_dr.shape
```

```python
plt.scatter(X_dr[:,0],X_dr[:,1],c=y)
plt.show()
# 发现样本不均衡
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
Kernel = ['linear','poly','rbf', 'sigmoid']
for k in Kernel:
    time0 = time()
    clf = SVC(kernel=k
              ,gamma='auto'
              ,degree=1  #默认是3，'poly'计算量太大
              ,cache_size=5000).fit(X_train,y_train)
    print('The accuracy under {} is {}'.format(k,clf.score(X_test,y_test)))
    print(datetime.datetime.fromtimestamp(time()-time0).strftime("%M:%S:%f"))
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
data = pd.DataFrame(x)
data.describe([0.01,0.05,0.1,0.25,0.5,0.75,0.9,0.99]).T
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
Kernel = ['linear','poly','rbf', 'sigmoid']
for k in Kernel:
    time0 = time()
    clf = SVC(kernel=k
              ,gamma='auto'
              ,degree=1
              ,cache_size=5000).fit(X_train,y_train)
    print('The accuracy under {} is {}'.format(k,clf.score(X_test,y_test)))
    print(datetime.datetime.fromtimestamp(time()-time0).strftime("%M:%S:%f"))
```

# SVC建议先做无量纲化，尤其是rbf和sigmoid，都不擅长原始数据

```python
# 选择rbf，继续调参
# 重要参数gamma： 只有核函数有
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
score = []
gamma_range = np.logspace(-10,1,50) #np.logspace(起始值，终止值，个数)  返回在对数刻度上均匀间隔的数字
for g in gamma_range:
    clf = SVC(kernel='rbf'
              ,gamma=g
              ,cache_size=5000).fit(X_train,y_train)
    score.append(clf.score(X_test,y_test))
print(max(score),gamma_range[score.index(max(score))])
    
plt.plot(np.logspace(-10,1,50), score)
plt.show()
```

```python
# poly  网格搜索调整gamma和coef0
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV

time0 = time()

gamma_range = np.logspace(-10,1,50)
coef0_range = np.linspace(0,5,10)

param_grid = dict(gamma = gamma_range
                 ,coef0 = coef0_range)

cv = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=0)
grid = GridSearchCV(SVC(kernel='poly',degree=1,cache_size=5000)
                    ,param_grid=param_grid
                    ,cv=cv)
grid.fit(x,y)

print('The best parameters are {} witha  score of {}'.format(grid.best_params_, grid.best_score_))
print(datetime.datetime.fromtimestamp(time()-time0).strftime("%M:%S:%f"))
# 反而效果变不好
```

# 重要参数C：用于权衡将“训练样本的正确分类”与“决策函数的边际最大化”的效力
# 类似正则化的惩罚项，可以 调整模型复杂度
# 浮点数，默认1,必须大于0
# C越大，那SVC会选择边际较小，能更好拟合训练集的决策边际。不过训练时间会更长，模型更加复杂

```python
# rbf 
score = []
c_range = np.linspace(4,7,50)
for c in c_range:
    clf = SVC(kernel='rbf'
              ,C=c
              ,cache_size=5000).fit(X_train,y_train)
    score.append(clf.score(X_test,y_test))
print(max(score),c_range[score.index(max(score))])
    
plt.plot(c_range, score)
plt.show()
```

```python
# linear
score = []
c_range = np.linspace(0.01,8,50)
for c in c_range:
    clf = SVC(kernel='linear'
              ,C=c
              ,cache_size=5000).fit(X_train,y_train)
    score.append(clf.score(X_test,y_test))
print(max(score),c_range[score.index(max(score))])
    
plt.plot(c_range, score)
plt.show()
```

```python
clf = SVC(kernel='rbf'
          , C=5.530612244897959
          , gamma=0.05
          , cache_size=5000).fit(X_train,y_train)
print("Train score: {}".format(clf.score(X_train, y_train)))
print("Test score: {}".format(clf.score(X_test, y_test)))
```
