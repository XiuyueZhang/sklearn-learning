```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
y = iris.target
x = iris.data
```

```python
# pca
pca = PCA(n_components=2)
pca = pca.fit(x)
x_dr = pca.transform(x)
```

```python
x_dr
```

```python
plt.figure()
color = ['black','orange','yellow']
for i in range(0,3):
    plt.scatter(x_dr[y == i,0],x_dr[y == i,1],c=color[i],label=iris.target_names[i])
plt.legend()
plt.title('PCA od Iris dataset')
plt.show()
```

累计可解释方差贡献率曲线：一条以降维后保留特征个数为横坐标，降维后新特征矩阵捕捉到的可解释方差贡献率为纵坐标

```python
pca_line = PCA().fit(x)
```

```python
pca_line.explained_variance_ratio_
```

```python
import numpy as np
np.cumsum(pca_line.explained_variance_ratio_)
```

```python
plt.figure()
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4])
plt.xlabel('number of components after dimension reduction')
plt.ylabel('cumulative explained variance ratio')
plt.show()
```

```python
#mle 最大拟然估计自然超参数
pca_mle = PCA(n_components='mle')
x_mle = pca_mle.fit_transform(x)
```

```python
x_mle
```

```python
pca_mle.explained_variance_ratio_.sum()
```

按信息占比选择超参数：[0,1]之间的浮点数，并且让参数svd_solver='full',表示希望降为后的总解释性方差占比。
首先尝试0.9，0.8

```python
pca_f = PCA(n_components=0.97,svd_solver='full')
x_f = pca_f.fit_transform(x)
```

```python
pca_f.explained_variance_ratio_
```

svd参数：奇异值分解
svd可以直接找出新的特征矩阵，所以计算速度快很多
奇异值越大，带有的信息量越大

```python
import numpy as np
import pandas as pd
data = pd.read_csv('train.csv')
```

```python
data.shape
```

```python
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

```python
data.head()
```

```python
y = data.iloc[:,0]
X = data.iloc[:,1:]
```

```python
X.shape
```

```python
#画累计方差贡献率曲线，找到最佳降维后的维度范围
pca_line = PCA().fit(X)
plt.figure(figsize=[20,5])
plt.plot(np.cumsum(pca_line.explained_variance_ratio_))
plt.xlabel('number of components after decomposition')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
#随机森林 学习曲线  找出最佳n_compoments
score = []
for i in range(15,25):
    pca = PCA(n_components=i).fit(X)
    X_dr = pca.transform(X)
    once = cross_val_score(RFC(n_estimators=10,random_state=0),X_dr,y,cv=5).mean()
    score.append(once)
plt.figure(figsize=[20,5])
plt.plot(range(15,25),score)
plt.show()
```

```python
score = []
for i in range(10,25,1):
    pca = PCA(n_components=i).fit(X)
    X_dr = pca.transform(X)
    once = cross_val_score(RFC(n_estimators=10,random_state=0),X_dr,y,cv=5).mean()
    score.append(once)
plt.figure(figsize=[20,5])
plt.plot(range(10,25,1),score)
plt.show()
```

```python
X_dr = PCA(n_components=21).fit_transform(X)
```

```python
X_dr.shape
```

```python
cross_val_score(RFC(n_estimators=10,random_state=0),X_dr,y,cv=5).mean()
```

```python
# 调整n_estimators 越大应该越好
cross_val_score(RFC(n_estimators=100,random_state=0),X_dr,y,cv=5).mean()
```

```python
# 再想办法：1. 调整随机森林  2. 换模型  特征数量只有21，可以用knn
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
# 默认k=5
cross_val_score(KNN(),X_dr,y,cv=5).mean()
```

```python
# knn学习曲线，找到最佳k值
score = []
for i in range(10):
    pca = PCA(n_components=21).fit_transform(X)
    once = cross_val_score(KNN(i+1),X_dr,y,cv=5).mean()
    score.append(once)
plt.figure(figsize=[20,5])
plt.plot(range(10),score)
plt.show()
```

```python
%%timeit
cross_val_score(KNN(3),X_dr,y,cv=5).mean()
```

```python
# 重要属性 components_
from sklearn.datasets import fetch_lfw_people
```

```python
faces = fetch_lfw_people(min_faces_per_person=60)
```

```python
faces
```

```python
faces.data.shape  #行：样本  列：样本相关的特征  每个图像62*47
```

```python
faces.images.shape  #图像数据的特征矩阵 - 三维数组
# 1348 图像个数 （一张图就是一个表）
# 62 每个图像的特征矩阵的行
# 47 每个图像的特征矩阵的列
```

```python
X = faces.data
X.shape
```

```python
# 原特征矩阵就是图像 可视化
fig, axes = plt.subplots(3,8   #4行5列，20张图
                        ,figsize=(8,4)  #比列，尺寸
                        ,subplot_kw = {'xticks':[],'yticks':[]} #不要显示坐标轴
                        )
# fig是整张图  axes是每个子图对象
# 提取images的每一个子图
# axes[0][0].imshow(faces,images[,:,:])  第一个子图填充画布
for i,ax in enumerate(axes.flat):
    ax.imshow(faces.images[i,:,:],cmap='gray')  #ax是每一个子图，颜色从绿色变成灰色
```

```python
axes.shape
```

```python
# 提取images的每一个子图
# axes[0][0].imshow(faces,images[,:,:])  第一个子图填充画布
for i,ax in enumerate(axes.flat):
    ax.imshow(faces,images[i,:,:],cmap='gray')  #ax是每一个子图，颜色从绿色变成灰色
```

```python
#axes降维到一维
len([*axes.flat])
enumerate(axes.flat)
```

```python
# PCA 降维
pca = PCA(150).fit(X)  #只接受二维矩阵
```

```python
V = pca.components_ 
```

```python
V.shape  #V(k,n) 这是用来映射新特征矩阵的向量空间
```

```python
# V(k,n) 画图
fig, axes = plt.subplots(3,8   #4行5列，20张图
                        ,figsize=(8,4)  #比列，尺寸
                        ,subplot_kw = {'xticks':[],'yticks':[]} #不要显示坐标轴
                        )
for i,ax in enumerate(axes.flat):
    ax.imshow(V[i,:].reshape(62,47),cmap='gray')  #ax是每一个子图，颜色从绿色变成灰色
```

```python
V[0,:].reshape(62,47)
```

```python
X_dr = pca.transform(X)
X_dr.shape
```

```python
# 降维后的数据X_dr 画图
fig, axes = plt.subplots(3,8   #4行5列，20张图
                        ,figsize=(8,4)  #比列，尺寸
                        ,subplot_kw = {'xticks':[],'yticks':[]} #不要显示坐标轴
                        )
for i,ax in enumerate(axes.flat):
    ax.imshow(X_dr[i,:].reshape(5,30),cmap='gray')  #ax是每一个子图，颜色从绿色变成灰色
```

```python
# 重要接口 inverse_transform： 还原原特征矩阵（实际上只是升维，并不会还原信息）- pca降维不可逆
X_dr.shape
```

```python
X_inverse = pca.inverse_transform(X_dr)
```

```python
X_inverse.shape
```

```python
# 降维后的数据X_dr 画图
fig, axes = plt.subplots(3,8   #3行8列，24张图
                        ,figsize=(8,4)  #比列，尺寸
                        ,subplot_kw = {'xticks':[],'yticks':[]} #不要显示坐标轴
                        )
for i,ax in enumerate(axes.flat):
    ax.imshow(X_inverse[i,:].reshape(62,47),cmap='gray')  #ax是每一个子图，颜色从绿色变成灰色
```

```python
# 对比pca降维前后的数据生成的图像
fig, axes = plt.subplots(2,10   
                        ,figsize=(10,2.5)  #比列，尺寸
                        ,subplot_kw = {'xticks':[],'yticks':[]} #不要显示坐标轴
                        )
for i in range(10):
    axes[0,i].imshow(faces.images[i,:,:],cmap='gray')
    axes[1,i].imshow(X_inverse[i,:].reshape(62,47),cmap='gray')
```

```python
# inverse_transform只是将特征空间升维，但是丢失的信息并不能找回
```
