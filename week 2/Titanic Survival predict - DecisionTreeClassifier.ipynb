```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv(r'C:\Users\le\Downloads\titanic\train.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
# Preprocessing data
data.drop(['Name','Cabin','Ticket'],inplace=True, axis=1)
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data = data.dropna()
```

```python
labels = data['Embarked'].unique().tolist()
data['Embarked'] = data['Embarked'].apply(lambda x: labels.index(x))
```

```python
data['Sex'] = (data['Sex']=='male').astype('int')   #二分类使用的方法  data.loc[:,'Sex']=='male'
```

```python
# feature, label分离
x = data.iloc[:,data.columns != 'Survived']
y = data.iloc[:,data.columns == 'Survived']
```

```python
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
X_train  #索引变乱，需要纠正索引
```

```python
 for i in (X_train, X_test, y_train, y_test):
        i.index = range(i.shape[0])
```

```python
X_train
```

```python
clf = DecisionTreeClassifier(random_state=25)
clf = clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)

score
```

```python
from sklearn.model_selection import cross_val_score

clf = DecisionTreeClassifier(random_state=25)
score = cross_val_score(clf, x, y, cv=10).mean()

score
```

```python
tr = []
te = []

for i in range(10):
    clf = DecisionTreeClassifier(random_state=25
                                ,max_depth=i+1
                                ,criterion='entropy')
    clf = clf.fit(X_train, y_train)
    score_tr = clf.score(X_train, y_train)
    score_te = cross_val_score(clf, x, y, cv=10).mean()
    tr.append(score_tr)
    te.append(score_te)
print(max(te))
plt.plot(range(1,11), tr, color='red', label='train')
plt.plot(range(1,11), te, color='blue', label='test')
plt.xticks(range(1,11))
plt.legend()
plt.show()
```

```python
import numpy as np

# 网格搜索  同时调整多个参数，枚举技术
parameters = {'criterion':('gini','entropy')
             ,'splitter':('best','random')
             ,'max_depth':[*range(1,10)]
             ,'min_samples_leaf':[*range(1,50,5)]
             ,'min_impurity_decrease':[*np.linspace(0,0.5,20)]
             }
```

```python
clf = DecisionTreeClassifier(random_state=25)
GS = GridSearchCV(clf, parameters, cv=10)
GS = GS.fit(X_train, y_train)
```

```python
GS.best_params_  #查看最佳参数
```

```python
GS.best_score_   #网格搜索后的模型评判标准
```
