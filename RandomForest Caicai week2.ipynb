```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_wine
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3)
```

```python
rfc = RandomForestClassifier(n_estimators=71,random_state=90)
rfc = rfc.fit(X_train, y_train)
rfc_s = rfc.score(X_test, y_test)

dtc = DecisionTreeClassifier(random_state=0)
dtc = dtc.fit(X_train, y_train)
dtc_s = dtc.score(X_test, y_test)
```

```python
rfc_s
```

```python
print("random forest score: {}".format(rfc_s), "single tree score: {}".format(dtc_s))
```

```python
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

```python
rfc = RandomForestClassifier(n_estimators=65)
rfc_s = cross_val_score(rfc, wine.data, wine.target, cv=10)

dtc = DecisionTreeClassifier()
dtc_s = cross_val_score(dtc,wine.data,wine.target,cv=10)


plt.plot(range(1,11),rfc_s,label='Random Forest')
plt.plot(range(1,11),dtc_s,label='Desicion Tree')
plt.legend()
plt.show()
```

```python
rfc_l = []
dtc_l = []

for i in range(10):
    rfc = RandomForestClassifier(n_estimators=65)
    rfc_s = cross_val_score(rfc, wine.data, wine.target, cv=10).mean()
    rfc_l.append(rfc_s)

    dtc = DecisionTreeClassifier()
    dtc_s = cross_val_score(dtc,wine.data,wine.target,cv=10).mean()
    dtc_l.append(dtc_s)


plt.plot(range(1,11),rfc_l,label='Random Forest')
plt.plot(range(1,11),dtc_l,label='Desicion Tree')
plt.legend()
plt.show()
```

```python
# n_estimators学习曲线

score_l=[]
for i in range(0,200,10):
    rfc = RandomForestClassifier(n_estimators=i+1,
                                n_jobs=-1,
                                random_state=90)
    score = cross_val_score(rfc,wine.data,wine.target,cv=10).mean()
    score_l.append(score)
print(max(score_l),(score_l.index(max(score_l))*10)+1)
plt.figure(figsize=[20,5])
plt.plot(range(1,201,10),score_l)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_boston
```

```python
df = load_boston()
```

```python
df.target
```

```python
df = pd.DataFrame(df.data)
df.head()
```

```python
x_full = df.data
y_full = df.target
n_samples = x_full.shape[0]
n_features = x_full.shape[1]
```

```python
n_samples
```

```python
n_features
```

```python
# 创造缺失值
# 确定缺失值比例：30%
rng = np.random.RandomState(0)
missing_rate = 0.3
n_missing_samples = int(np.floor(n_samples*n_features*missing_rate))

# 一个缺失的数据需要一个行索引和一个列索引
missing_features = rng.randint(0,n_features,n_missing_samples)
missing_samples = rng.randint(0,n_samples,n_missing_samples)
```

```python
x_missing = x_full.copy()
y_missing = y_full.copy()

# 生成的随机索引的那些值，有NaN替代
x_missing[missing_samples, missing_features] = np.nan
x_missing = pd.DataFrame(x_missing)
```

```python
x_missing
```

```python
from sklearn.impute import SimpleImputer
# 用0填补
x_0 = SimpleImputer(missing_values=np.nan,strategy='constant',fill_value=0).fit_transform(x_missing)
```

```python
x_0 = pd.DataFrame(x_0)
```

```python
# 用mean填补
x_mean = SimpleImputer(missing_values=np.nan,strategy='mean').fit_transform(x_missing)
```

```python
x_mean = pd.DataFrame(x_mean)
```

```python
# 用median填补
x_median = SimpleImputer(missing_values=np.nan,strategy='median').fit_transform(x_missing)
x_median = pd.DataFrame(x_median)
```

```python
x_median.isnull().sum()
```

```python
# 随机森林回归填补缺失值
x_missing.head()
x_missing_reg = x_missing.copy()
```

```python
# 从小到大排序的顺序所对应的索引
sortindex = np.argsort(x_missing.isnull().sum()).values
```

```python
from sklearn.ensemble import RandomForestRegressor as rfr

for i in sortindex:
    # 构建新的矩阵和标签
    df = x_missing_reg
    # 新标签
    fillc = df.iloc[:,i]
    # 新矩阵：其他列+原标签列y_full
    df = pd.concat([df.iloc[:,df.columns!=i],pd.DataFrame(y_full)],axis=1)
    
    # 在新的矩阵中，对含有缺失值的列，用0进行填补
    df_0 = SimpleImputer(missing_values=np.nan,strategy='constant',fill_value=0).fit_transform(df)
    
        
    # 找出训练集和测试集
    y_train = fillc[fillc.notnull()]
    y_test = fillc[fillc.isnull()]
    
    x_train = df_0[y_train.index,:]
    x_test = df_0[y_test.index,:]
    
    #RFR填补缺失值
    rfc = rfr(n_estimators=100)
    rfc = rfc.fit(x_train, y_train)
    y_predict = rfc.predict(x_test)
    
    #填补好的特征返回原特征矩阵
    x_missing_reg.loc[x_missing_reg.iloc[:,i].isnull(),i] = y_predict
 
```

```python
x_missing_reg.isnull().sum()
```

```python
from sklearn.model_selection import cross_val_score

x = [x_0,x_mean,x_median,x_missing_reg,x_full]
mse = []
std = []
for i in x:
    estimator = rfr(random_state=0, n_estimators=100)
    scores = cross_val_score(estimator, i, y_full,scoring='neg_mean_squared_error',cv=5).mean()
    mse.append(scores*(-1))
```

```python
mse #mse越小越好
```

```python
[*zip(['x_0,x_mean','x_median','x_missing_reg','x_full'],mse)]
```

```python
# plot
import matplotlib.pyplot as plt

x_labels = ['zero imputation','mean imputation','median imputation','Random Forest imputation','Full data']
colors = ['r','b','g','orange','gray']

plt.figure(figsize=(12,6))
ax = plt.subplot(111)
for i in range(len(mse)):
    ax.barh(i,mse[i],color=colors[i],alpha=0.6,align='center')
ax.set_title('Imputation Techniques with Boston Data')
ax.set_xlim(left=np.min(mse)*0.9, right=np.max(mse)*1.1)  #x轴的刻度
ax.set_yticks(np.arange(len(mse)))
ax.set_xlabel('MSE')
ax.set_yticklabels(x_labels)
plt.show()
```
